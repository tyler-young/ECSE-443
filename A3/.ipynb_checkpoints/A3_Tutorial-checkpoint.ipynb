# Using np.pad

* For the assignment always use **mode='median'** as shown below

```python
# Generate a 2D signal
x = 0.00001 * np.arange(1920*1080)
signal2d = np.abs(np.sin(x)).reshape(1080, 1920)
signal2d[:, 700:1220] = 0
plt.imshow(signal2d)
plt.title('A 2D signal')
plt.show()

# pad an the signal, note the use of mode='median'
top = 500
bottom = 0
left = 0
right = 0 
signal2dPad = np.pad(signal2d, ((top,bottom), (left,right)), mode='median')
plt.imshow(signal2dPad)
plt.title('2D signal - top padding')
plt.show()

# bottom padding
top = 0
bottom = 500
left = 0
right = 0 
signal2dPad = np.pad(signal2d, ((top,bottom), (left,right)), mode='median')
plt.imshow(signal2dPad)
plt.title('2D signal - bottom padding')
plt.show()

# left padding
top = 0
bottom = 0
left = 500
right = 0 
signal2dPad = np.pad(signal2d, ((top,bottom), (left,right)), mode='median')
plt.imshow(signal2dPad)
plt.title('2D signal - left padding')
plt.show()

# right padding
top = 0
bottom = 0
left = 0
right = 500 
signal2dPad = np.pad(signal2d, ((top,bottom), (left,right)), mode='median')
plt.imshow(signal2dPad)
plt.title('2D signal - right padding')
plt.show()
```

# 1D convolution

```python
# define a filter, reshape into a row vector
filter1d = (np.sin(0.2 * np.arange(17))).reshape(1,-1)
filter1d = filter1d / np.sum(filter1d)
plt.plot(filter1d.T)
plt.title("A 1D filter")
plt.show()

# define a signal, reshape into a row vector
signal1d = np.abs((np.sin(200 * np.arange(100)) * 10)).reshape(1, -1)
plt.plot(signal1d.T)
plt.title("A 1D signal")
plt.show()

# find filter radius
radius1d = filter1d.shape[1] // 2 # Note: x // 2 is same as int(x / 2)

# pad the 1D signal
signal1dPad = np.pad(signal1d, ((0,0), (radius1d,radius1d)), mode='median')
plt.plot(signal1dPad.T)
plt.title("Padded 1D signal")
plt.show()

# create another array to store the result of 1D convolution
result = np.zeros(signal1d.shape)
signalLength = result.shape[1]

# 1D convolution algorithm
for i in range(radius1d, radius1d + signalLength):
  # compute weighted sum
  # notice the use of vectorization to compute a weighted sum
  result[0, i - radius1d] = np.sum(signal1dPad[0, i - radius1d : i + radius1d + 1] * filter1d)

plt.plot(result.T)
plt.title("Result of 1D convolution")
plt.show()
```

# Spatially varying 1D convolution

* Regular convolution - Size/Radius of the filter remains same at each index (in 1D) or pixel (in 2D) of the input signal.
* Spatially varying convolution - Size/Radius of the filter varies at each index (in 1D) or pixel (in 2D) of the input signal.
* Regular convolution - uses a single filter.
* Spatially varying convolution - uses a stack of filters from radius 1 all the way up to maximum (provided as input to the problem).



```python
# define a stack of filters
maxRadius = 8
stack1d = []
for radius in range(1, maxRadius +1):
  filter1d = np.abs(np.arange(-radius, radius +1))
  filter1d = filter1d / np.sum(filter1d)
  stack1d.append(filter1d)

# define an input signal
signal1d = np.abs((np.sin(200 * np.arange(100)) * 10)).reshape(1, -1)
plt.plot(signal1dPad.T)
plt.title("Padded 1D signal")
plt.show()

# A function to query the radius at a given index, an equivalent of this function is provided to you in A3
# You can use this function as a black box to query the radius at i^th index
def radiusOracle(index):
  return (index % maxRadius) + 1

# Similar to regular 1D case, pad the signal with maxRadius
signal1dPad = np.pad(signal1d, ((0,0), (maxRadius,maxRadius)), mode='median')
plt.plot(signal1dPad.T)
plt.title("Padded 1D signal")
plt.show()

# create another array to store the result of spatially varying 1D convolution
result = np.zeros(signal1d.shape)
signalLength = result.shape[1]

# spatially varying 1D convolution algorithm
for i in range(maxRadius, maxRadius + signalLength):
  # query the radius of filter at a given index i
  radius1d = radiusOracle(i - maxRadius)
  # look-up into the stack using radius as key to obtain the filter
  filter1d = stack1d[radius1d - 1]
  # compute weighted sum
  result[0, i - maxRadius] = np.sum(signal1dPad[0, i - radius1d : i + radius1d + 1] * filter1d)

plt.plot(result.T)
plt.title("Result of spatially varying 1D convolution")
plt.show()
```

# 2D convolution

```python
# Generate a 2D signal
x = 0.0001 * np.arange(1280*720)
signal2d = np.abs(np.sin(x) + np.random.rand(1280*720)).reshape(720, 1280)
plt.imshow(signal2d)
plt.title('A 2D signal')
plt.show()

# Generate a convolution filter
filter1d = (np.sin(0.1 * np.arange(35))).reshape(1,-1)
filter2d = np.outer(filter1d, filter1d) 
filter2d = filter2d / np.sum(filter2d)
plt.imshow(filter2d)
plt.title("A 2D filter")
plt.show()

# find the radius of filter
radius2d = filter2d.shape[0] // 2

# Pad the input signal
signal2dPad = np.pad(signal2d, ((radius2d, radius2d),(radius2d, radius2d)), mode='median')
plt.imshow(signal2dPad)
plt.title("Padded 2D signal")
plt.show()

# create an array to store the result of 2d convolution
result2d = np.zeros(signal2d.shape)

# 2d convolution algorithm
height, width = signal2d.shape
for y in range(radius2d, radius2d + height):
  for x in range(radius2d, radius2d + width):
    result2d[y - radius2d, x - radius2d] = np.sum(signal2dPad[y - radius2d : y + radius2d + 1, x - radius2d : x + radius2d + 1] * filter2d)

plt.imshow(result2d)
plt.title("Result of 2D convolution")
plt.show()
```
